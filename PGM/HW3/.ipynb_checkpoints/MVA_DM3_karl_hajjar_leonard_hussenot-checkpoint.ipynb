```python
def log_Gaussian_density(x, mus, sigmas):
    #print(sigmas.shape)
    return - np.log(2*Pi) - 0.5 * np.log(np.linalg.det(sigmas))  - 0.5 * np.dot((x-mus).T, 
                                                                               np.dot(np.linalg.inv(sigmas), (x-mus))
                                                                              ) 
```

```python
def compute_log_conditionals(data, mus, sigmas, K=4):
    T = len(data)
    log_p = np.zeros([T,K])
    for k in range(K):
        log_p[:,k] = np.apply_along_axis(lambda x: log_Gaussian_density(x, mus[k], sigmas[k]), axis=1, arr=data)
    return log_p
```

```python
def check_sums(arr):
    ## Assumes arr is the log of some array, which we which to check if the sums of the ligns are equal to 1 using the 
    ## log-sum-exp trick
    shape = arr.shape
    if len(shape) == 1:
        max_ = np.max(arr)
        tmp = arr - max_
        log_sum = max_ + np.log(np.sum(np.exp(tmp)))
        res_sum = np.exp(log_sum)
    
    elif len(shape) == 2:
        max_ = np.max(arr, axis=1).reshape((-1,1))
        tmp = arr - max_
        log_sum = max_+ np.log(np.sum(np.exp(tmp), axis=1)).reshape((-1,1))
        res_sum = np.exp(log_sum)

    else:
        res_sum = np.sum(np.exp(arr), axis=(1,2))
        
    return res_sum
```

```python
def k_means(data, k=4, plot =True):
    
    precision = 10e-2
    n = len(data)
    n_iter = 50
    centers = data[np.random.choice(range(n), k, replace=False),:]
    old_centers= np.copy(centers)
    clusters = -1*np.ones((n,1))
    distances = np.zeros((n,k))
    
    keep_going = True
    iteration = 0
    while keep_going:
        iteration +=1
        #compute distances
        for i in range(n):
            for j in range(k):
                distances[i,j] = np.sum((data[i,:]-centers[j,:])**2)

        #compute clusters
        clusters  = np.argmin(distances, 1)

        #update centroids
        for j in range(k):

            centers[j,:] = np.mean(data[clusters==j], 0)
        
        v = centers - old_centers
        
        error = np.trace(np.dot(v,v.T))
        keep_going = error > precision
        old_centers = np.copy(centers)

            
    distance_to_centroids = np.min(distances, 1)
    distortion = np.sum(distance_to_centroids)

    if plot:
        label_color_map = {0 : 'r',
                           1 : 'g',
                           2 : 'b',
                           3 : 'purple'
                           }

        label_color = [label_color_map[l] for l in clusters]
        plt.scatter(data[:,0], data[:,1], c=label_color, marker="o", s=50)
        plt.scatter(centers[:,0], centers[:,1], color="yellow", marker='v', s=150)
        plt.xlabel("x")
        plt.ylabel("y")
        plt.title("K-means")
        plt.show()
        
    return distortion, distances, centers, clusters
```

```python
k=4
precision = 10e-1
data = u_train
n = len(data)
#Initialization with K-means
_, distances, centers, clusters = k_means(data, k, plot=False)

mu = centers
pi = np.array([len(clusters[clusters==j]) for j in range(k)])*(1/len(data))
sigma_2 = np.array([np.mean(np.sum((data[clusters==j]-centers[j])**2,1)) for j in range(k)])

old_mu = np.copy(mu)
old_pi = np.copy(pi)
old_sigma_2 = np.copy(sigma_2)

keep_going = True
iteration = 0
while keep_going :
    iteration +=1
    #E-step
    tau = np.zeros((n,k))
    p_x_z = (1/2*np.pi)*np.ones((n,k))
    for i in range(n):
        for j in range(k):
            p_x_z[i,j] *= (1/sigma_2[j])*np.exp((-0.5/sigma_2[j])*np.dot(data[i,:]-mu[j,:],
                                                                        (data[i,:]-mu[j,:]).T))

    for i in range(n):
        for j in range(k):
            tau[i,j] = p_x_z[i,j] * pi[j] / np.dot(p_x_z[i,:],pi.T)

    #M-step

    pi = np.mean(tau,0)
    a = np.sum(tau,0)
    mu = np.dot(tau.T,data)/np.array([a,a]).T

    for j in range(k):
        dists = np.array([np.sum((data[i,:]-mu[j,:])**2) for i in range(n)])
        sigma_2[j] = 0.5*np.dot(tau[:,j], dists) /np.sum(tau[:,j])
        
    #termination condition
    error = max(np.max(np.abs(mu-old_mu)), np.max(np.abs(pi-old_pi)), np.max(np.abs(sigma_2-old_sigma_2))) 
    keep_going = error > precision
    
    old_pi = np.copy(pi)
    old_mu = np.copy(mu)
    old_sigma_2 = np.copy(sigma_2)
```

```python
def alpha_beta_recursion(data, log_pi, log_A, mus, sigmas, K=4):
    
    assert(len(log_pi) == len(mus) == len(sigmas) == K)
    
    T = n_observ = data.shape[0]
    
    log_alpha = np.zeros([T,K])
    log_beta = np.zeros([T,K])
    
    log_p = compute_log_conditionals(data, mus, sigmas, K)
    log_alpha[0,:] = log_pi
    
    for t in range(T-1):
        ## alpha-recursion : using log-sum-exp trick
        log_alpha_t = log_alpha[t,:].reshape(-1,1)
        tmp = log_A + log_alpha_t
        maxs = np.max(tmp, axis=0)
        log_alpha[t+1,:] = log_p[t+1,:] + maxs + np.log(np.sum(np.exp(tmp - maxs), 
                                                               axis=0)
                                                        )
    for t in range(T-2,-1, -1):
        ## beta-recursion : using logs same as before
        log_beta_t_plus_1 = log_beta[t+1,:].reshape(-1,1)
        log_p_t_plus_1 = log_p[t+1,:].reshape(-1,1)
        tmp = log_A + log_beta_t_plus_1 + log_p_t_plus_1
        maxs = np.max(tmp, axis=0)
        log_beta[t,:] = maxs + np.log(np.sum(np.exp(tmp - maxs),
                                             axis=0)
                                     )
    
    return log_alpha, log_beta
```

```python
def compute_log_smoothing(log_alpha, log_beta):
    assert(len(log_alpha) == len(log_beta))
    T, K = log_alpha.shape
    
    log_smoothing = log_alpha + log_beta
    
    tmp = log_alpha + log_beta
    max_tmp = np.max(tmp, axis=1).reshape(-1,1)
    log_evidence = max_tmp + np.log(np.sum(np.exp(tmp - max_tmp), axis=1)).reshape(-1,1) 
    
    res = log_smoothing - log_evidence
    #assert ((check_sums(res) - 1.) < 1e-6).all()
    return res
```

```python
def compute_log_double_smoothing(log_alpha, log_beta, log_A, log_p):
    assert(len(log_alpha) == len(log_beta))
    T, K = log_alpha.shape
    log_double_smoothing = []
    
    tmp = log_alpha + log_beta
    max_tmp = np.max(tmp, axis=1).reshape(-1,1)
    log_evidence = max_tmp + np.log(np.sum(np.exp(tmp - max_tmp), axis=1)).reshape((-1,1)) 
    
    for t in range(T-1):
        #log_num = log_alpha[t,:].reshape((1,-1)) + log_beta[t+1,:] + log_A + log_p[t+1,:]  
        log_num = log_alpha[t,:].reshape((1,-1)) + log_beta[t+1,:].reshape(-1,1) + log_A + log_p[t+1,:].reshape((-1,1))  
        add = log_num - log_evidence[t,0]
        log_double_smoothing.append(add)
    
    res = np.array(log_double_smoothing)
    
    print(res.shape)
    #assert ((check_sums(res) - 1.) < 1e-6).all()
    return res
```

```python
def EM(data, log_pi, log_A, mus, sigmas, n_iter, data_test, K=4):
    
    T = len(data)
    llhs = []
    
    for it in tqdm(range(n_iter)):
        # E-step:
        log_alpha, log_beta = alpha_beta_recursion(data, log_pi, log_A, mus, sigmas, K)
        log_p = compute_log_conditionals(data, mus, sigmas)
        Edti = compute_log_smoothing(log_alpha, log_beta)
        Edtij = compute_log_double_smoothing(log_alpha, log_beta, log_A, log_p)
        
        #assert ((np.sum(Edtij, axis=2) - Edti[:T-1,:]) < 1e-6).all()

        
        # M-step
        
        ## log_pi
        new_log_pi = Edti[0,:]
        #print('min_diffs(new_log_pi, log_pi) = {}'.format(min_diffs(new_log_pi, log_pi)))
        log_pi = np.copy(new_log_pi)
        
        ## log_A
        maxs_num = np.max(Edtij, axis=0)
        log_num = maxs_num + np.log(np.sum(np.exp(Edtij - maxs_num), axis=0))
        
        log_den = np.sum(log_num, axis=1)
        
        
        #pprint(log_num)
        #pprint(np.sum(log_num, axis=1))
        #pprint(log_den)
        
        #assert ((np.sum(log_num, axis=1) - log_den.reshape((K,))) < 1e-6).all()
        
        #Edti_T_minus_1 = Edti[:T-1,:]
        #maxs_den = np.max(Edti_T_minus_1, axis=0)
        #log_den = maxs_den + np.log(np.sum(np.exp(Edti_T_minus_1 - maxs_den), axis=0))
        
        new_log_A = log_num - log_den
        #assert ((check_sums(new_log_A) - 1.) < 1e-3).all()
        #print('min_diffs(new_log_A, log_A) = {}'.format(min_diffs(new_log_A, log_A)))
        log_A = np.copy(new_log_A)
        
        
        ## mus
        exp_Edti = np.exp(Edti)
        renorm = np.sum(exp_Edti, axis=0).reshape(-1,1)
        new_mus = np.dot(exp_Edti.T, data) / renorm
        #print('min_diffs(new_mus, mus) = {}'.format(min_diffs(new_mus, mus)))
        mus = np.copy(new_mus)
        
        
        ## sigmas
        new_sigmas = []
        for k in range(4):
            sigma = np.zeros((2,2))
            for t in range(T):
                sigma += exp_Edti[t,k]*np.dot((data[t,:]-mus[k,:]).reshape(-1,1),(data[t,:]-mus[k,:]).reshape(1,-1))
            new_sigmas.append(sigma / renorm[k])
            
        for k in range(K):
            #print('min_diffs(new_sigmas[k], sigmas[k]) = {}'.format(min_diffs(new_sigmas[k], sigmas[k])))
            sigmas[k] = np.copy(new_sigmas[k])
        
        #new_log_alpha, new_log_beta = alpha_beta_recursion(data, log_pi, log_A, mus, sigmas)
        #new_log_p = compute_log_conditionals(data, mus, sigmas)
        #new_Edti = compute_log_smoothing(new_log_alpha, new_log_beta)
        #new_Edtij = compute_log_double_smoothing(new_log_alpha, new_log_beta, log_A, new_log_p)
        
        #print('np.min(np.abs(new_log_alpha - log_alpha)) = {}'.format(min_diffs(new_log_alpha, log_alpha)))
        #print('np.min(np.abs(new_log_beta - log_beta)) = {}'.format(min_diffs(new_log_beta, log_beta)))
        #print('np.min(np.abs(new_log_p - log_p)) = {}'.format(min_diffs(new_log_p,log_p)))
        #print('min_diffs(new_Edti, Edti) = {}'.format(min_diffs(new_Edti, Edti)))
        #print('min_diffs(new_Edtij, Edtij) = {}'.format(min_diffs(new_Edtij, Edtij)))
        
        #print('')
        

        # Compute log-likelihood
        llh = np.dot(np.exp(log_pi),log_pi)
        llh += np.sum(np.exp(Edtij)*(log_A))
        llh += np.sum(np.exp(Edti)*log_p)

        llhs.append(llh)
        
    clusters = np.argmax(Edti,1)
    final_llh = 0
    for i in range(len(clusters)):
        final_llh += log_Gaussian_density(data[i],mus[clusters[i]],sigmas[clusters[i]]) 
    print("final log_likelihood on train data is : {}".format(final_llh))
    
    
    log_alpha, log_beta = alpha_beta_recursion(data_test, log_pi, log_A, mus, sigmas, K)
    log_p = compute_log_conditionals(data_test, mus, sigmas)
    Edti = compute_log_smoothing(log_alpha, log_beta)
    
    clusters = np.argmax(Edti,1)
    final_llh = 0
    for i in range(len(clusters)):
        final_llh += log_Gaussian_density(data_test[i],mus[clusters[i]],sigmas[clusters[i]]) 
    print("final log_likelihood on test data is : {}".format(final_llh))
     
    return log_pi, log_A, mus, sigmas, llhs
```

As you can see, we now obtain a better likelihood on both our training and testing data.

It makes sense to make the comparison between the Gaussian mixtures model and the HMM because they are both versions of an EM algorithm which tries to maximise the likelihood of the observations $p(\bar{y} \ | \ \theta)$ with different sets of parameters. Thus the likelihoods obtained during both EM algorithms are lower bounds on the true likelihood of the observed data.

However, in fact, the likelihoods $p(\bar{y} \ | \ \theta)$ are not computed under the same distribution $p$ for both models since for the HMM, the Markov chain implies coupling between $z_{t+1}$ and $z_t$ whereas in the Gaussian Mixtures, the value of $z_{t+1}$ is independent of $z_t$.

```python
## Viterbi algorithm
# Uses the learned parameters logA, log_pi etc...

def viterbi(data, log_pi, log_A, mus, sigmas, T_end): #Compute most probable sequence until T_end
    log_p = compute_log_conditionals(data, mus, sigmas)

    messages = np.zeros((T,K))
    messages[0,:] = log_pi + log_p[0,:]

    for t in range(1,T_end):
        vec = np.max(log_p[t,:].reshape(1,-1) + log_A + messages[t-1,:].reshape(-1,1), axis=0)
        messages[t,:] = vec

    most_probable_sequence = []
    #probas = []

    most_probable_sequence.append(np.argmax(messages[T_end-1,:]))
    #probas.append(np.max(messages[T_end-1,:]))
    for t in range(2,T_end+1):
        tmp = log_p[T_end-t,:].reshape(1,-1) + log_A + messages[T_end-t+1,:].reshape(-1,1)
        vec = tmp[:,most_probable_sequence[-1]]
        i_realized_the_max = np.argmax(vec)
        
        most_probable_sequence.append(i_realized_the_max)
    
    return most_probable_sequence[::-1]
```

```python
def plot_clutering(data, log_pi, log_A, mus, sigmas, d_type='train'):

    clusters = viterbi(data, log_pi, log_A, mus, sigmas, len(data))
    
    final_llh = 0
    for i in range(len(clusters)):
        final_llh += log_Gaussian_density(data[i],mus[clusters[i]],sigmas[clusters[i]]) 
    print("final log_likelihood on train data is : {}".format(final_llh))
    
    label_color_map = {0 : 'r',
                       1 : 'g',
                       2 : 'b',
                       3 : 'y'
                       }

    label_color = [label_color_map[l] for l in clusters]


    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, aspect='equal')



    ax.scatter(data[:,0], data[:,1], c=label_color, marker="o", s=60, label="datapoints")
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    
    plt.scatter(mus[:,0], mus[:,1], color="k", marker="H", s=200, label="cluster centers")
    plt.legend()
    
    plt.title('Max-Product algorithm to predict states on {}ing set'.format(d_type))
    plt.show()
```

```python
def plot_most_probable_states(most_probable_states, d_type='train', K=4):
    
    states_by_index = [np.argwhere(most_probable_states==k).reshape((-1,)) for k in range(K)]
    
    colors = ['red', 'green', 'purple', 'blue']
    plt.figure(1, figsize=(18,7))
    for k in range(K):
        indexes = states_by_index[k]
        plt.scatter(indexes, [k+1]*len(indexes), color=colors[k], label = 'state {}'.format(k+1))
    plt.title('Most probable states for arriving data points of {}ing set'.format(d_type))
    plt.xlabel('t')
    plt.ylabel('predicted state')
    plt.xlim(0,100)
    plt.ylim(0.7,4.1)
    plt.legend()
    plt.show()
```

```python
log_alpha, log_beta = alpha_beta_recursion(data, log_pi, log_A, mus, sigmas)
log_smoothing = compute_log_smoothing(log_alpha, log_beta)
most_probable_states = np.argmax(log_smoothing, axis=1)[:100]
plot_most_probable_states(most_probable_states, d_type='test')
```
